```python
import os
import subprocess

os.chdir("/content")  # default Colab working dir

urls = [
    "https://huggingface.co/comfyanonymous/flux_text_encoders/resolve/main/t5xxl_fp16.safetensors?download=true",
    "https://huggingface.co/Lightricks/LTX-Video/resolve/main/ltx-video-2b-v0.9.safetensors?download=true",
]

for url in urls:
    # derive a clean filename from the URL
    fname = url.split("/")[-1].split("?")[0]
    print(f"Downloading {fname} ...")
    subprocess.run(
        ["wget", "-O", fname, url],
        check=True
    )

print("All downloads finished. Files are in /content")
```

```python
import shutil
import os

# Define where your ComfyUI is located (usually /content/ComfyUI)
base_path = "/content/ComfyUI/models"

# Mapping: { "filename": "destination_folder" }
move_map = {
    "t5xxl_fp16.safetensors": os.path.join(base_path, "clip"),
    "ltx-video-2b-v0.9.safetensors": os.path.join(base_path, "checkpoints")
}

for fname, dest_folder in move_map.items():
    source = os.path.join("/content", fname)
    destination = os.path.join(dest_folder, fname)

    if os.path.exists(source):
        # Create folder if it doesn't exist (e.g., 'clip' might not exist in some installs)
        os.makedirs(dest_folder, exist_ok=True)

        print(f"Moving {fname} to {dest_folder}...")
        shutil.move(source, destination)
    else:
        print(f"Skipping {fname}: File not found in /content. Did the download fail?")

print("All files moved successfully!")
```

```python
import os
from google.colab import files

output_folder = "/content/ComfyUI/output"

all_files = []

# Walk through all subfolders
for root, dirs, files_in_dir in os.walk(output_folder):
    for file in files_in_dir:
        path = os.path.join(root, file)
        all_files.append(path)

# Print all files
print("Found files:\n")
for f in all_files:
    print(f)

# Download them
print("\nDownloading files...\n")
for f in all_files:
    files.download(f)
```
